## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, import `Orders` from Ironhack's database into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here
orders = pd.read_csv('C:/Users/gixi_/Downloads/Orders.csv')
orders
```

```python
# your code here
customer_spent = orders.groupby('CustomerID', as_index=False).agg({'amount_spent':'sum'})
vip_customer = customer_spent.loc[customer_spent['amount_spent'] >= customer_spent.amount_spent.quantile(0.95)]
preferred_customer = customer_spent.loc[(customer_spent['amount_spent'] >= customer_spent.amount_spent.quantile(0.75)) & (customer_spent['amount_spent'] < customer_spent.amount_spent.quantile(0.95))]
```

## Q2: How to identify which country has the most VIP Customers?

```python
vip_country = pd.merge(vip_customer, orders, on='CustomerID').groupby('Country',as_index=False).agg({'CustomerID':'nunique'})
vip_country.sort_values('CustomerID', ascending = False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
```

```python
preferred_country = pd.merge(preferred_customer, orders, on='CustomerID').groupby('Country',as_index=False).agg({'CustomerID':'nunique'})
preferred_country.sort_values('CustomerID', ascending = False)
```

```python
country = pd.merge(vip_country, preferred_country, on='Country')
country
```
