```python
#avg salary per company, top 20

avg_salary_company = df_subset.groupby('company').agg({'basesalary':'mean','stockgrantvalue':'mean', 'bonus':'mean','totalyearlycompensation':'mean'})
avg_salary_company.reset_index(inplace = True)
avg_salary_company.sort_values('totalyearlycompensation', ascending = False).head(20)
```

```python
#avg salary per title 
avg_salary_title = df_subset.groupby('title').agg({'basesalary':'mean','stockgrantvalue':'mean', 'bonus':'mean','totalyearlycompensation':'mean'})
avg_salary_title.reset_index(inplace = True)
avg_salary_title.sort_values('totalyearlycompensation', ascending = False)
```

```python
#avg salary per title per company
avg_salary_title_company = df_subset.groupby(['company','title']).agg({'totalyearlycompensation':['mean','max','min']})
avg_salary_title_company.columns = ['totalyearlycompensation_mean', 'totalyearlycompensation_max', 'totalyearlycompensation_min']
avg_salary_title_company.reset_index(inplace = True)
avg_salary_title_company.sort_values('totalyearlycompensation_max', ascending = False).head(20)
```
