```python
import pandas as pd
import numpy as np
```

```python
dict = {'TransactionID': (1,2,3,4,5),
       'ClientID': (231,765,453,231,892),
        'Profession': ('Self-employed','students','Horeca','self-employed','finance'),
        'Bank_dep': ('009', '005', '007', '009', '003'),
        'Risk' : ('High', 'High', 'Medium', 'High','Low'),
        'Number of credits': (2,3,5,2,3),
        'Revenue':(30200,12700,89400,30200,740000)
       }
df = pd.DataFrame(dict)
df
```

```python
# Get dummies method

gd = pd.get_dummies(df)
gd
```

```python
# Label encoding

categorical_cols = ['Profession', 'Risk']

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df[categorical_cols] = df[categorical_cols].apply(lambda col: le.fit_transform(col))
df 
```

```python
# One Hot encoding

from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe.fit(df)
onehotlabel = ohe.transform(df).toarray()
df_onehotlabel = pd.DataFrame(onehotlabel)
df_onehotlabel
```

```python
# Ordinal encoding

from sklearn.preprocessing import OrdinalEncoder

risk_column = df[['Risk']]

encoder = OrdinalEncoder()
risk_encoded = encoder.fit_transform(risk_column)
risk_encoded
df_risk_encoded = pd.DataFrame(risk_encoded)
df_risk_encoded.rename(columns = {0:'Risk'}, inplace=True)
df_risk_encoded
```
