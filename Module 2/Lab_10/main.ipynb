#### How many customers do we have in Spain?

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
df1 = orders.groupby('CustomerID')['Quantity'].agg({'count'})
df1.loc[df1['count']>50]
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here
df2 = orders.groupby('InvoiceNo')['Quantity'].agg('sum').to_frame()
df2.loc[df2['Quantity']>50]
```

#### Select all free orders.

```python
# your code here
df3 = orders.groupby('InvoiceNo')['amount_spent'].agg('sum').to_frame()
df3.loc[df3['amount_spent']==0]
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
# your code here
orders.loc[orders['Description'].str.startswith('lunch bag')==True]
```

#### Select all `lunch bag` orders made in 2011.

```python
# your code here
orders.loc[(orders['Description'].str.startswith('lunch bag')==True) & orders['InvoiceDate'].str.startswith('2011')]
```

#### Show the frequency distribution of the amount spent in Spain.

```python
# your code here
import matplotlib.pyplot as plt
df4 = orders.loc[orders['Country']=='Spain']
plt.hist(df4['amount_spent'], bins = 50)
plt.xlabel('Amount Spent')
plt.ylabel('Frequencies') 
plt.title('Frequency Distribution of the Amount Spent in Spain')
plt.grid()
plt.show()
```
