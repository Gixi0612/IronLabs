#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
Minutes_Lightly_Active = fitbit["Minutes Lightly Active"]
Minutes_Very_Actives = fitbit['Minutes Very Active']
Minutes_Sedentary= fitbit['Minutes Sedentary']
Steps= fitbit['Steps'] 
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.scatter(Steps, Minutes_Lightly_Active, color='r')
ax.scatter(Steps, Minutes_Very_Actives, color='b')
ax.scatter(Steps, Minutes_Sedentary, color='g')
ax.set_title('scatter plot')
ax.set_xlabel('Steps')
ax.set_ylabel('Minutes')
plt.show()
```
